```python
import csv
import pandas as pd

file_path = './data/osdg.csv'

df = pd.read_csv(file_path, sep='\t')
df
```

```python
df_osdg = df[['text', 'sdg']]

df_osdg
```

```python
df_osdg.groupby('sdg').count()
```

- osdg 데이터에는 SDG-17이 존재하지 않으므로, datahub의 데이터 중 SDG-17에 해당하는 데이터를 추가

```python
df_hub = pd.read_csv('data_hub.csv')
df_hub.head()
```

```python
sdg = [f'SDG-{str(i).zfill(2)}' for i in range(1, 18)]

columns_to_select = ['text'] + sdg
df_hub = df_hub[columns_to_select]

df_hub[sdg].sum()
```

```python
sdg_17_only = df_hub[df_hub['SDG-17'] == 1]

other_sdgs = [sdg for sdg in sdg if sdg != 'SDG-17']
sdg_17_plus_one_other = sdg_17_only[other_sdgs].sum(axis=1) <= 1

sdg_17_plus_one_other_count = sdg_17_plus_one_other.sum()
sdg_17_plus_one_other_count
```

```python
sdg_17_plus_one_other_rows = sdg_17_only[sdg_17_plus_one_other]
sdg_17_plus_one_other_rows
```

```python
# SDG-17==1
sdg_17_only = df_hub[df_hub['SDG-17'] == 1]

# other sdgs sum==0
sdg_sum_zero = sdg_17_only[other_sdgs].sum(axis=1) == 0
sdg_sum_zero_rows = sdg_17_only[sdg_sum_zero]

# other sdgs sum==1
sdg_sum_one = sdg_17_only[other_sdgs].sum(axis=1) == 1
sdg_sum_one_rows = sdg_17_only[sdg_sum_one]

sdg_sum_zero_rows
```

SDG-17에만 해당하는 데이터: 616개

```python
sdg_sum_one_rows
```

다른 SDG 중 하나를 포함하고, SDG-17==1을 만족하는 데이터: 803개

```python
remaining_rows_count = 1100 - len(sdg_sum_zero_rows)

if remaining_rows_count > 0:
    selected_sdg_sum_one_rows = sdg_sum_one_rows.head(remaining_rows_count)
else:
    selected_sdg_sum_one_rows = pd.DataFrame()

# DataFrame concat
final_rows = pd.concat([sdg_sum_zero_rows, selected_sdg_sum_one_rows])
final_rows
```

osdg 데이터에서 SDG 별로 존재하는 최소 데이터는 약 1100개 이므로 총 1100개의 SDG-17 데이터셋 정립

```python
df_osdg
```

```python
final_rows_to_add = final_rows[['text']].copy()  
final_rows_to_add['sdg'] = 17 

df_osdg_updated = pd.concat([df_osdg, final_rows_to_add], ignore_index=True)
df_osdg_updated
```
